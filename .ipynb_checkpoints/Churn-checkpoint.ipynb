```python
#converting yes or no columns into integer datatype
#1 for "yes" and 0 for "no"
# Set the future downcasting behavior
#FutureWarning: Downcasting behavior in replace is deprecated and will be removed in a future version. To retain the old behavior, explicitly call result.infer_objects(copy=False). To opt-in to the future behavior, set pd.set_option('future.no_silent_downcasting', True)
df = df.replace({'Yes': 1, 'No': 0})
pd.set_option('future.no_silent_downcasting', True)
df = df.replace({'Yes': 1, 'No': 0})
```

```python
df
```

```python
churned_customers = df[df['Churn']== 1]
churn_rate = len(churned_customers)/ len(df)* 100
print("churn rate = ", churn_rate, "%")
```

```python
# Group by multiple columns and filter where 'Churn' is 1 (churned customers)
churned_customers
```

```python
# Create a summary of characteristics, such as gender, partner, etc.
characteristics = churned_customers.groupby([ 'gender']).size().reset_index(name='Count')
```

```python
# Display ordered characteristics by count of churned customers
characteristics_sorted = characteristics.sort_values(by='Count', ascending=False)
```

```python
characteristics_sorted
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=churned_customers, order=churned_customers['gender'].value_counts().index)
plt.title('Gender Distribution of Churned Customers')
plt.xlabel('Gender')
plt.ylabel('Count')
# plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of characteristics to analyze
characteristics_list = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 
                        'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 
                        'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 
                        'Contract', 'PaperlessBilling', 'PaymentMethod']

# Loop through each characteristic
for characteristic in characteristics_list:
    # Group by the characteristic and count the number of churned customers
    characteristics = churned_customers.groupby([characteristic]).size().reset_index(name='Count')
    
    # Sort by count in descending order
    characteristics_sorted = characteristics.sort_values(by='Count', ascending=False)
    
    # Print the sorted characteristics for reference
    print(f"\n{characteristic} distribution among churned customers:\n")
    print(characteristics_sorted)
    
    # Plot a bar chart for the characteristic
    plt.figure(figsize=(8, 6))
    sns.countplot(x=characteristic, data=churned_customers, order=churned_customers[characteristic].value_counts().index)
    plt.title(f'{characteristic} distribution among churned customers')
    plt.xlabel(characteristic)
    plt.ylabel('Count')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.show()

# For numerical characteristics like 'tenure', 'MonthlyCharges', 'TotalCharges', we will plot histograms
numerical_characteristics = ['tenure', 'MonthlyCharges', 'TotalCharges']

for num_char in numerical_characteristics:
    plt.figure(figsize=(8, 6))
    sns.histplot(churned_customers[num_char], bins=10, kde=False)
    plt.title(f'{num_char} distribution among churned customers')
    plt.xlabel(num_char)
    plt.ylabel('Count')
    plt.show()
```
